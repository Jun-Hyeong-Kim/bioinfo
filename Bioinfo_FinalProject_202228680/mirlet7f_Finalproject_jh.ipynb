```python
import pandas as pd

pileup = pd.read_csv('CLIP-let7f-gene.pileup', sep='\t', names=['chrom', 'pos', '_ref', 'count', 'basereads', 'quals'])
pileup.tail()
```

```python
import re
toremove = re.compile('[<>$*#^]')
pileup['matches'] = pileup['basereads'].apply(lambda x: toremove.sub('', x))
```

```python
pileup[['chrom', 'pos', 'matches']]
```

```python
import math 
shannon = []
for pos in pileup['pos'].unique():
  lowercase_count = sum(pileup[pileup['pos'] == pos]['matches'].str.count('[a-z]'))
  lowercase_count_a = sum(pileup[pileup['pos'] == pos]['matches'].str.count('a'))
  lowercase_count_c = sum(pileup[pileup['pos'] == pos]['matches'].str.count('c'))
  lowercase_count_g = sum(pileup[pileup['pos'] == pos]['matches'].str.count('g'))
  lowercase_count_t = sum(pileup[pileup['pos'] == pos]['matches'].str.count('t'))
  a_prob = lowercase_count_a/lowercase_count
  c_prob = lowercase_count_c/lowercase_count
  g_prob = lowercase_count_g/lowercase_count
  t_prob = lowercase_count_t/lowercase_count
  prob = [a_prob, c_prob, g_prob, t_prob]
  shannon.append(sum([-base_prob*math.log2(base_prob) for base_prob in prob if base_prob != 0]))
```

```python
shannon
```

```python
import pandas as pd

# Create a new dataframe with 'chrom', 'pos', and 'end' columns
bedgraph = pd.DataFrame({'chrom': pileup['chrom'], 'pos': pileup['pos']-1, 'end': pileup['pos']}) # bedgraph index 0  -1 

# Add the 'shannon' column from the 'shannon' list
bedgraph['shannon'] = shannon

bedgraph
```
